```python
import os
import json
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import numpy as np
import pandas as pd
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
import joblib
import h5py
```

```python
file_path = 'D:\\Dataset\\#4_       \\Training\\1._\\1.\\1.\\0.Combined\\'
files = ['Combined_LabelledData_52_.json', 'Combined_LabelledData_56_.json']

x_label = []
y_label = []

for file_name in files:
    full_path = file_path + file_name
    with open(full_path, 'r', encoding='utf-8') as file:
        data = json.load(file)

    for entry in data['data']:
        if entry['ITEM_NAME'] == "":
            x_label.append(entry['ITEM_VALUE'])
            # y_label.append(entry['LABEL_NAME'])
            # Loading -> 0
            if entry['LABEL_NAME']=='Loading':
                y_label.append(0)
            # Unloaidng -> 1
            else : 
                y_label.append(1)
            

x_label = np.array(x_label)

print(x_label)
print(y_label)

```

```python
x_label = np.array(x_label).reshape(-1, 1)
y_label = np.array(y_label)

#  train test 
X_train, X_test, y_train, y_test = train_test_split(x_label, y_label, test_size=0.2, random_state=42)

# SVM   
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# SVM    
svm_model = SVC(kernel='rbf', gamma='scale')  # rbf   SVM 
svm_model.fit(X_train, y_train)

#   
y_pred = svm_model.predict(X_test)

#  
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.4f}')
```

```python
def train_and_save_model(x_train, y_train, model_save_path):
    #  
    joblib.dump(svm_model, model_save_path)

model_save_path = 'C:\\Users\\user\\Desktop\\test\\Define\\Define_add\\svm_model.pkl'
train_and_save_model(X_train, y_train, model_save_path)
```
