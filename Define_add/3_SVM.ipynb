```python
import os
import json
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import numpy as np
import pandas as pd
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
import joblib
import h5py
```

```python
file_path = 'D:\\Dataset\\#4_전력 설비 에너지 패턴 및 고장 분석 센서\\Training\\1.펌프_일반모터\\1.기동패턴\\1.전류평균\\0.Combined\\'
files = ['Combined_LabelledData_52_전류평균.json', 'Combined_LabelledData_56_전류평균.json']

x_label = []
y_label = []

for file_name in files:
    full_path = file_path + file_name
    with open(full_path, 'r', encoding='utf-8') as file:
        data = json.load(file)

    for entry in data['data']:
        if entry['ITEM_NAME'] == "전류평균":
            x_label.append(entry['ITEM_VALUE'])
            # y_label.append(entry['LABEL_NAME'])
            # Loading -> 0
            if entry['LABEL_NAME']=='Loading':
                y_label.append(0)
            # Unloaidng -> 1
            else : 
                y_label.append(1)
            

x_label = np.array(x_label)

print(x_label)
print(y_label)

```

```python
x_label = np.array(x_label).reshape(-1, 1)
y_label = np.array(y_label)

# 데이터를 train과 test로 나눔
X_train, X_test, y_train, y_test = train_test_split(x_label, y_label, test_size=0.2, random_state=42)

# SVM에 사용할 데이터를 정규화
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# SVM 모델 생성 및 학습
svm_model = SVC(kernel='rbf', gamma='scale')  # rbf 커널을 사용한 SVM 모델
svm_model.fit(X_train, y_train)

# 테스트 데이터로 예측
y_pred = svm_model.predict(X_test)

# 정확도 출력
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.4f}')
```

```python
def train_and_save_model(x_train, y_train, model_save_path):
    # 모델 저장
    joblib.dump(svm_model, model_save_path)

model_save_path = 'C:\\Users\\user\\Desktop\\test\\Define\\Define_add\\svm_model.pkl'
train_and_save_model(X_train, y_train, model_save_path)
```

```python
# 모델 로드
loaded_model = joblib.load('C:\\Users\\user\\Desktop\\test\\Define\\Define_add\\svm_model.pkl')  # 모델 파일 경로로 변경

# 테스트 데이터에 대한 예측 수행
y_pred = loaded_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.4f}')

# Confusion matrix 출력
cm = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:')
print(cm)

# Classification report 출력
classification_rep = classification_report(y_test, y_pred)
print('Classification Report:')
print(classification_rep)

# 문자열을 숫자로 변환 (예시: '1'을 1로, '0'을 0으로)
y_pred_numeric = y_pred.astype(float)

# 이후에 비교 연산을 수행
y_pred_binary = (y_pred_numeric > 0.5)
print("Binary Predictions:")
print(y_pred_binary)
```
